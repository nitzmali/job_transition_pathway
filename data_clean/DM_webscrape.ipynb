```python
import requests
import time
import json

# Define the base URL for the Adzuna API
BASE_URL = "https://api.adzuna.com/v1/api/jobs/us/search/1?app_id=912bf73d&app_key=bf52616f3833157430b32c135ec2190f&what_phrase=data%20engineer"
OUTPUT_FILE = "/Users/nyzy/Library/CloudStorage/GoogleDrive-nitinmali999@gmail.com/.shortcut-targets-by-id/1yEbTjc1DwwTDd2CB86Il6F-3DhFUn8pU/Data Mining/data/data_engineer.json"

APP_ID = "912bf73d"
APP_KEY = "bf52616f3833157430b32c135ec2190f"

params = {
    "app_id": APP_ID,
    "app_key": APP_KEY,
    "results_per_page": 50  # max number of results per page
}

all_jobs = []

page_num = 1
fetch = True
while fetch:
    print(f"Fetching page {page_num}...")
    
    #response = requests.get(BASE_URL + str(page_num), params=params)
    response = requests.get(BASE_URL)
    
    if response.status_code != 200:
        print(f"Failed to fetch page {page_num}. Stopping.")
        break

    data = response.json()
    if not data['results']:
        print("No more results found. Stopping.")
        break

    all_jobs.extend(data['results'])

    # Pause for a short while to avoid hitting rate limits
    time.sleep(2)
    
    page_num += 1
    fetch = False

# Save all the fetched job data to a JSON file
with open(OUTPUT_FILE, "w") as file:
    json.dump(all_jobs, file)


print(f"Finished fetching job postings. Data saved to {OUTPUT_FILE}.")
```

```python
import requests
from bs4 import BeautifulSoup
import time

BASE_URL = "https://www.indeed.com/jobs?l=United+States&start="

page_number = 0
all_jobs = []

while True:
    response = requests.get(BASE_URL + str(page_number * 10))
    if response.status_code != 200:
        print("Error fetching the page.")
        break
    
    soup = BeautifulSoup(response.content, "html.parser")
    job_cards = soup.find_all("div", class_="jobsearch-SerpJobCard")
    
    if not job_cards:  # No more jobs found
        break

    for card in job_cards:
        job_title = card.find("h2", class_="title").find("a").get("title")
        company = card.find("span", class_="company")
        if company:
            company_name = company.get_text(strip=True)
        else:
            company_name = "N/A"
        all_jobs.append({"Job Title": job_title, "Company": company_name})

    page_number += 1
    time.sleep(5)  # Sleep for 5 seconds to avoid too many requests

print(f"Found {len(all_jobs)} job listings.")
```
